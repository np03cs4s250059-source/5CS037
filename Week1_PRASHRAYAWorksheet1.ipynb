```python
#Problem - 4: Matrix Operations:

#For the following arrays:
#x = np.array([[1,2],[3,5]]) and y = np.array([[5,6],[7,8]]);
#v = np.array([9,10]) and w = np.array([11,12]);
#Complete all the task using numpy:

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

#1)  A.A*−1 = I.
A_inv = np.linalg.inv(A)
print("1. A * A^-1:\n", np.dot(A, A_inv))

#2) AB ̸= BA.
AB = np.dot(A,B)
BA = np.dot(B,A)
print("\n2. AB:\n", AB)
print("\n   BA:\n", BA)
print("\n   Clearly AB != BA")

#3) (AB)T = BTAT
lhs = AB.T
rhs = np.dot(B.T , A.T)
print("\n3. (AB)^T:\n", lhs)
print("\n   B^T A^T:\n", rhs)





```

```python
#Solve the following system of Linear equation using Inverse Methods.

#2x − 3y + z = −1
#x − y + 2z = −3
#3x + y − z = 9

C = np.array([
    [2, -3,  1],
    [1, -1,  2],
    [3,  1, -1]
])

D = np.array([-1, -3, 9])

C_inv = np.linalg.inv(C)
solution = np.dot(C_inv, D)

print("\nSolution using inverse:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])

solution2 = np.linalg.solve(C, D)
print("\nSolution using np.linalg.solve:\n", solution2)
```
